1. Implement a phone book as a dictionary.

```python
Phone_book = {"John":"123 452 600","George":"123 854 203","Alice":"1234 203 157", "John S":"123 027 589", "Joseph":"1234 568 751","John Pi":"123 574 693"}

Phone_book
```

```python
file_path='../Course 1/'

def save_phone_book(Phone_book, file_path):
    with open(file_path, 'w') as file:
        for name, number in Phone_book.items():
            file.write(f"{name}: {number}\n")

# Save phone book to file
save_phone_book(Phone_book, "Phone_book.txt")
```

2. Find the most frequent words in a text file using sets.

```python
def most_frequent_word(file_path):
    word_counts = {}
    with open(file_path, 'r') as file:
        words = file.read().split()
        for word in words:
            word_counts[word] = word_counts.get(word, 0) + 1
    sorted_words = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)
    most_common_word, count = sorted_words[0]
    return most_common_word, count

# Example usage
most_common_word, count = most_frequent_word("Phone_book.txt")
print(f"Most common word: {most_common_word} and it counts: {count}")
```

3. Filter user-submitted comments stored in a list for profanity.

```python
profanity_filter = {"bad", "offensive", "inappropriate"}

def filter_comments(comments):
    filtered_comments = []
    for comment in comments:
        words = comment.lower().split()
        if not any(word in profanity_filter for word in words):
            filtered_comments.append(comment)
    return filtered_comments

# Example usage
comments = [
    "This is a good comment.",
    "That was a bad comment.",
    "I found the language offensive in that comment.",
    "Such an inappropriate comment!"
]
```

4. Build a histogram from word counts using dictionaries.

```python
import matplotlib.pyplot as plt

def word_histogram(file_path):
    word_counts = {}
    with open(file_path, 'r') as file:
        words = file.read().split()
        for word in words:
            word_counts[word] = word_counts.get(word, 0) + 1
    return word_counts

histogram = word_histogram('Phone_book.txt')

# Plotting the histogram
plt.bar(histogram.keys(), histogram.values())
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Word Histogram')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

5. Convert a deeply nested list to a flat list.

```python
def flatten_list(nested_list):
    flat_list = []
    for item in nested_list:
        if isinstance(item, list):
            flat_list.extend(flatten_list(item))
        else:
            flat_list.append(item)
    return flat_list

# Example usage
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
flat_list = flatten_list(nested_list)
print(flat_list)
```
